```python
print(np.where(adit['SepalWidthCm']>4))
print(np.where(adit['SepalWidthCm']<2.1))
```

```python
#IQR
Q1 = np.percentile(adit['SepalWidthCm'], 25, interpolation= 'midpoint')
Q3 = np.percentile(adit['SepalWidthCm'], 75, interpolation= 'midpoint')
IQR = Q3-Q1
print(IQR)
```

```python
print('Old Shape: ', adit.shape)
```

```python
upper = np.where(adit['SepalWidthCm']>= (Q3+1.5*IQR))
lower = np.where(adit['SepalWidthCm']<= (Q1-1.5*IQR))
```

```python
adit.drop(upper[0], inplace = True)
adit.drop(lower[0], inplace = True)

print('New Shape: ', adit.shape)
```
